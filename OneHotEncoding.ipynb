# OneHotEncoding

A method of turning categorical data into boolean data

```python
import pandas as pd
data = ['male','male','male','male','male','female','male','male','female','male','male','male','male','female','female']

gender_df = pd.DataFrame(data, columns = ['Gender'])
gender_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()
label_encoded = le.fit_transform(gender_df.Gender)
print(label_encoded)
gender_encodeddf = pd.DataFrame(label_encoded , columns = ['GenderEncoded']) 
gender_encodeddf.head()
gender_encodeddf
```

```python
from sklearn.preprocessing import OneHotEncoder

# creating an instance of the class
ohe = OneHotEncoder()
one_encoded =ohe.fit_transform(gender_encodeddf[['GenderEncoded']])
one_encodeddf = pd.DataFrame(one_encoded, columns = ['Gender_OneHotEncoded'])
one_encodeddf
```

```python
# creating dummies as an alternative to onehotencoding
dummies = pd.get_dummies(gender_encodeddf['GenderEncoded'])
dummiesdf = pd.DataFrame(dummies)
dummiesdf
concatenated = pd.concat([gender_df,dummiesdf],axis = 1)
concatenated
```
